```python
import scipy.io
import numpy as np
import librosa
import os
from python_speech_features import mfcc
from scipy.signal import resample
```

```python
# Function to load .mat files
def load_mat_file(file_path):
    data = scipy.io.loadmat(file_path)
    return data
```

```python
# Function to extract MFCCs using librosa
def extract_mfcc_librosa(signal, sr, n_mfcc=13):
    mfccs = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=n_mfcc)
    return mfccs
```

```python
# Function to extract MFCCs using python_speech_features
def extract_mfcc_psf(signal, sr, n_mfcc=13):
    mfccs = mfcc(signal, sr, numcep=n_mfcc)
    return mfccs
```

```python
# Function to preprocess the signal (e.g., resample, normalize)
def preprocess_signal(signal, target_sr=16000):
    # Resample if necessary
    if target_sr != 12000:  # Assuming original sampling rate is 12 kHz
        signal = resample(signal, int(len(signal) * target_sr / 12000))
    # Normalize the signal
    signal = signal / np.max(np.abs(signal))
    return signal
```

```python
# Main function to process all .mat files in a directory
def process_mat_files(directory, output_dir, sr=16000, n_mfcc=13):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for file_name in os.listdir(directory):
        if file_name.endswith('.mat'):
            file_path = os.path.join(directory, file_name)
            data = load_mat_file(file_path)

            # Assuming the vibration signal is stored under a key like 'X097_DE_time'
            signal_key = [key for key in data.keys() if 'time' in key][0]
            signal = data[signal_key].flatten()

            # Preprocess the signal
            signal = preprocess_signal(signal, target_sr=sr)

            # Extract MFCCs (choose one method)
            mfccs = extract_mfcc_librosa(signal, sr=sr, n_mfcc=n_mfcc)  # Using librosa
            # mfccs = extract_mfcc_psf(signal, sr=sr, n_mfcc=n_mfcc)  # Using python_speech_features

            # Save the MFCCs
            output_file = os.path.join(output_dir, file_name.replace('.mat', '.npy'))
            np.save(output_file, mfccs)
            print(f"Processed and saved MFCCs for {file_name} to {output_file}")
```

```python
# Example usage
mat_files_directory = '/content/drive/MyDrive/FYP Phase II/Dataset FYP II/Normal  baseline data'
output_directory = '/content/drive/MyDrive/FYP Phase II/MFCC Files/Normal baseline data'
process_mat_files(mat_files_directory, output_directory)
```
